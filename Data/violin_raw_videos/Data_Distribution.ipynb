```python
import os
import glob
import math
import csv

def get_all_video_files(folder_path, extensions=['.mp4', '.avi', '.mkv', '.mov']):
    video_files = []
    for ext in extensions:
        video_files.extend(glob.glob(os.path.join(folder_path, f'*{ext}')))
    return video_files

folder_path = './clips'
video_files = get_all_video_files(folder_path)

for video_file in video_files:
    print(video_file)
```

```python
def split_list_into_chunks(lst, n):
    chunk_size = math.ceil(len(lst) / n)
    return [lst[i:i + chunk_size] for i in range(0, len(lst), chunk_size)]

def save_to_csv(files, output_filename):
    with open(output_filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['File Name'])  
        for file in files:
            writer.writerow([file])
```

```python
output_folder = './parts/'
chunks = split_list_into_chunks(video_files, 7)

# CSV
for i, chunk in enumerate(chunks):
    output_filename = os.path.join(output_folder, f'video_files_part_{i + 1}.csv')
    save_to_csv(chunk, output_filename)

```

```python
import json

transcripts = json.load(open('text_data.json', 'r'))
```

```python
#  split  'test'  id
test_ids = [file_id for file_id, info in transcripts.items() if info['split'] == 'test']

#  test ids
print(len(test_ids))
```

```python
premise_count = 0
hypo_count = 0

#  premise
for file_id in test_ids:
    info = transcripts[file_id]
    
    # file_idpremise
    premise_count += 1
    
    # hypostatementrealfake
    if 'statement' in info:
        hypo_count += len(info['statement']) * 2  # statementrealfake2

#  premise  hypo 
print(f"Total Premises in Test Set: {premise_count}")
print(f"Total Hypotheses in Test Set: {hypo_count}")
```
